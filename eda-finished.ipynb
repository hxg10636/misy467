## Task 1
There are 8 basic functions as described at http://mathonweb.com/help_ebook/html/functions_4.htm. Do the followings:
1. create 8 subplots (4x2 4 rows with two subplots each row) with figure size (15, 15)
2. plot a function (you choose) of each type in each subplot, add the function type as the title of the subplot, and the quadratic function has to be in the subplot in the second row and second column
3. use at least 4 colors, 4 different line styles, and 4 different markers for the 8 functions
4. change the x label to x and y label to the function
5. add legends for each subplot to show the functions, such as y = 3x +5

```python
# create 8 sublots
fig,ax = plt.subplots(4,2,figsize=(15,15))
# define x
x = np.linspace(-10, 10, 100)

# plot graph 1
ax[0,0].set_title('linear function',fontsize=15)
ax[0,0].plot(x,x+2, color='green',linestyle='-', marker='v')
ax[0,0].legend(['y=x+2'])
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y=x+2')

# plot graph 2
ax[0,1].set_title('power function',fontsize=15)
ax[0,1].plot(x,x**10, color='blue',linestyle='--', marker='o')
ax[0,1].legend(['y=x**10'])
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('y=x**10')

# plot graph 3
ax[1,0].set_title('polynomial function',fontsize=15)
ax[1,0].plot(x,(x**5)-8*(x**3)+10*x+6, color='yellow',linestyle='-.', marker='<')
ax[1,0].legend(['y=(x**5)-8*(x**3)+10*x+6'])
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y=(x**5)-8*(x**3)+10*x+6')

# plot graph 4
ax[1,1].set_title('quadratic function',fontsize=15)
y = x**2 + 2*x + 2  
ax[1,1].plot(x,y, color='gray',linestyle=':', marker='*')
ax[1,1].legend(['y=x**2 + 2*x + 2'])
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('y=x**2 + 2*x + 2')

# plot graph 5
ax[2,0].set_title('rational function',fontsize=15)
ax[2,0].plot((x**2+1)/(x))
ax[2,0].legend(['y=x**2+1'])
ax[2,0].set_xlabel('x')
ax[2,0].set_ylabel('y=x**2+1')

# plot graph 6
ax[2,1].set_title('exponential function',fontsize=15)
ax[2,1].plot((1/2)**x)
ax[2,1].legend(['y=(1/2)**x'])
ax[2,1].set_xlabel('x')
ax[2,1].set_ylabel('y=(1/2)**x')

# plot graph 7
ax[3,0].set_title('logarithmic function',fontsize=15)
ax[3,0].plot(np.log(x))
ax[3,0].legend(['y=log(x)'])
ax[3,0].set_xlabel('x')
ax[3,0].set_ylabel('y=log(x)')

# plot graph 8
ax[3,1].set_title('sinusoidal function',fontsize=15)
ax[3,1].plot(2*np.sin(3*x))
ax[3,1].legend(['y=2*sin(3x)'])
ax[3,1].set_xlabel('x')
ax[3,1].set_ylabel('y=2*sin(3x)')
```

- calculate the suvival rate using `value_counts()`
- you cannot hardcode any number

```python
# calculate the survival rate
a = df['Survived'].value_counts()
print(a)
print(f'the survival rate is {a[1]/a.sum():.1%}')
```

- create the histogram for gender
- calculate the survival rate based on gender using `groupby()`
- you cannot hardcode any number

```python
# create the histogram for gender.
df['Sex'].hist()
```

```python
# groupby survived passengers based on their gender.
group = df.groupby('Sex')['Survived'].value_counts()
# calculate the survival rate
print(f'female survival rate: {group["female", 1]/group["female"].sum():.1%}')
print(f'male survival rate: {group["male", 1]/group["male"].sum():.1%}')
```
