```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.style.use('seaborn')
```

```python
fig = plt.figure()
fig=plt.figure(figsize=(10,10))
ax = plt.axes()
x = np.linspace(0, 10, 50)
ax.plot(x,x)
ax.plot(x,x+3)
ax.plot(x,2*x+3)
fig.savefig('myimage.png')
```

```python
fig=plt.figure(figsize=(15,15))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,4)
ax1.plot(x,x*2)
ax2.plot(x,x**2)
```

```python
fig,ax = plt.subplots(2,2,figsize=(15,15))
ax[0,0].plot(x,x+2)
ax[1,0].plot(x,x**2)
```

```python
fig,ax  = plt.subplots(2,figsize=(15,15))
ax[0].plot(x,x+2)
ax[1].plot(x,x**2)
```

```python
fig,ax  = plt.subplots(figsize=(15,15))
ax.set_title('this is my title',fontsize=25)
ax.set_xlabel('this is x label',fontsize=25)
ax.set_ylabel('this is y label',fontsize=25)

ax.set_xlim(0,5)
ax.set_ylim(0,15)
ax.plot(x,x**2,label='y=x^2',color='green',linestyle='dashed')
ax.plot(x,x+5,label='y=x+5')
ax.legend(fontsize=20)
```
